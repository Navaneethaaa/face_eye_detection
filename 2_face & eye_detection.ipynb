```python

import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('data/index.jpeg')
```

```python

converted=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray)
```

```python
face_cascade=cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')
```

```python
faces=face_cascade.detectMultiScale(converted)
```

```python
print(faces)
```

```python
len(faces)
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(converted,(x,y),(x+w,y+h),color=(0,255,0),thickness=5)
```

```python
plt.imshow(converted)
```

```python
eye_cascade=cv2.CascadeClassifier('data/haarcascade_eye.xml')
```

```python
eyes=eye_cascade.detectMultiScale(converted)
```

```python
eyes
```

```python
for(x,y,w,h) in eyes:
    cv2.rectangle(converted,(x,y),(x+w,y+h),color=(0,255,0),thickness=1)
```

```python
plt.imshow(converted)
```
